```python
from Bio.SeqIO.FastaIO import SimpleFastaParser
prot_path = "hCoV-19_spikeprot0913/"
nuc_path = "hCoV-19_spikenuc0914"
file = open(prot_path + "spikeprot0913.fasta", errors="ignore")
virus_dict = dict()
n = 10
for title, seq in SimpleFastaParser(file):
    id = title.split(sep="|")[3]
    virus_dict[id] = seq
    # n -= 1
    # if n == 0:
    #     break
file.close()
virus_dict
```

```python
import csv
len(virus_dict)
file = open("omicron.csv")
omicron_id = list()
for row in csv.reader(file):
    omicron_id.append(row[0])
file.close()
```

```python
omicron_id = omicron_id[1:]
```

```python
omicron_dict = {key : virus_dict[key] for key in omicron_id if key in virus_dict}
len(omicron_dict)
```
